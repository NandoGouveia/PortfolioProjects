```python
#Import Libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Read in the data

df = pd.read_csv(r'C:\Users\nando\Downloads\movies.csv')
```

```python
# let's look at the data

df.head()
```

```python
#checking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#drop all missing data
df = df.dropna()
```

```python
#data types for our columns

df.dtypes
```

```python
#Changing data type of columns (removing decimals)

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
#turning the date into the correct format

#df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#scatter plot with budget vs gross (revenue)

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget')
plt.show()
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation matrix for numeric features')

plt.xlabel('Movie Features')

plt.ylabel('Movie features')

plt.show()
```

```python
# Looking at company

#we can represent a string in a numeric, than use correlation matri (only works with numbers)
df.head()
```

```python
#changes object datatype into category and give each data a unique number
# ex: (A specific company would have a unique number so it can appear >1 time)

df_numerized = df             

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype =='object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
                                   
                                   
df_numerized.head()
```

```python
# Now we can make a correlation Matrix
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation matrix for numeric features')

plt.xlabel('Movie Features')

plt.ylabel('Movie features')

plt.show()
```

```python
df_numerized.corr()
```

```python
#unstacks shows correlation in a form of a table but by a single column 
#compared to each other column

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs.head()
```

```python
# Shows correlation between two columns in pairs (ex: genre vs budget,.. ,..)

sorted_pairs = corr_pairs.sort_values()

sorted_pairs.head()
```

```python
# Shows all in which correlation is above 0.5

high_corr = sorted_pairs[(sorted_pairs)>0.5]

high_corr
```
